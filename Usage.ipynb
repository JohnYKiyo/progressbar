```python
import progressbar
import time
```

```python
prog = progressbar.progbar(10,clear_display=False)
for i in range(10):
    time.sleep(0.1)
    prog.update(i,info="hoge {t}".format(**dict(t=i)))
del prog
```

```python
prog = progressbar.progbar(10,clear_display=True)
for i in range(10):
    time.sleep(0.5)
    prog.update(i,info="hoge {t}".format(**dict(t=i)))
del prog
```
